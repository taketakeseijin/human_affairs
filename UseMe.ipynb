```
# 名簿データcsvを入れてください。
# 名前の列名は"name" , 所属の列名は "group" としてください。
df = pd.read_csv(f"{path}/test1.csv")
```

```
member_num = df.shape[0]
print(f"人数は{member_num}")
```

```
# 部署数設定。
department_num = 8
department_names = [f"D{i}" for i in range(department_num)]
```

```
# ターム数設定。
term_num = department_num
term_names = [f"T{i}" for i in range(term_num)]
```

```
# 所属ごとに配属される部署・ターム (region)を指定できます。

np_all = np.ones((department_num,term_num))
np_rest = np.eye(term_num)

# 所属数分regionの設定をしてください。
# ↓サンプルで所属が２の場合
np_A = np.zeros((department_num,term_num))
np_A[[2,3],0:2] = 1
np_A[[4,5],2:4] = 1
np_A[[6,7],4:6] = 1
np_A[[0,1],6:8] = 1
np_B = np_all - np_rest - np_A

region_dict = {}
region_dict["A"] = np_A
region_dict["B"] = np_B

# 設定したregionの表示
for key,value in region_dict.items():
  print("グループ："+key)
  print(value,end="\n\n")
```

↑データ入力
↓配属プログラムを作る

```
# 最適化の後に、よい結果を save_folderに出力。
save_folder = "test"
WOpt = wrapper.Opt(df,department_names,term_names,region_dict,path+"/"+save_folder)
```
